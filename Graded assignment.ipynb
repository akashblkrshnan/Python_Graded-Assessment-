```python
### e. What is the median class interval of age?

class_intervals = [(5,15),(15,25),(25,35),(35,45),(45,55),(55,65)]
frequencies = [6,11,21,23,14,5]
cumulative_frequencies = [6,17,38,61,75,80]

N = sum(frequencies)

N_by_2 = N/2

median_class_index = next(i for i,cf in enumerate(cumulative_frequencies)if cf > N_by_2)
median_class = class_intervals[median_class_index]

L  = median_class[0] 
CF = cumulative_frequencies[median_class_index - 1] if median_class_index > 0 else 0
f = frequencies[median_class_index]
h  = median_class[1] - median_class[0]

median = L + ((N_by_2 -CF) / f) * h


print(f'Median class interval :{median_class}')
print(f'Median : {median : .2f}')
        

```

```python
### 9. Evaluate the measures of central tendency and measures of dispersion for all the quantitative variables in the dataset.

import pandas as pd

df = pd.read_csv('Bank Personal Loan Modelling (1).csv')

quantitative_vars = ['Age','Experience','Income','CCAvg','Mortgage']

results = {}

for var in quantitative_vars :
    
    mean = df[var].mean()
    median = df[var].median()
    mode =  df[var].mode()[0] if not df[var].mode().empty else 'No mode'
    std_dev = df[var].std()
    variance = df[var].var()
    range = df[var].max() - df[var].min()
    iqr = df[var].quantile(0.75) -df[var].quantile(0.25)
    

results[var] = {
    'Mean': mean,
    'Median': median,
    'Mode': mode,
    'Standard Deviation' : std_dev,
    'Variance' : variance,
    'Range' : range,
    'Interquartile Range (IQR)' : iqr
}

for var,metrics in results.items():
    print(f'Variable: {var}')
    for metric,value in metrics.items():
        print(f' {metric}:  {value}')
        print()
 




    


```

```python
### 10. What statistical method will you use to examine the presence of a linear relationship between age and experience variables? 
### Also,create a plot to illustrate this relationship.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy .stats import linregress

df = pd.read_csv('Bank Personal Loan Modelling (1).csv')

correlation = df['Age'].corr(df['Experience'])
print(f'Correlation Coefficient between Age and Experience:{correlation:.2f}')

plt.figure(figsize=(10,6))
plt.scatter(df['Age'],df['Experience'],alpha=0.5)
plt.title('Scatter Plot of Age vs.Experience')
plt.xlabel('Age')
plt.ylabel('Experience')


slope,intercept,r_value,p_value,std_err = linregress(df['Age'], df['Experience'])
plt.plot(df['Age'], intercept + slope * df['Age'], 'r', label =f'Fit line: y={intercept:2f}+{slope:.2f}x')
plt.legend()

plt.show()

print(f'Slope:{slope:.2f}')
print(f'Intercept: {intercept:.2f}')
print(f'R-squared; {r_value**2:.2f}')








```

```python
""" Answer : To examine the presence of a linear regression between age and experience variables we could use the Correlation coefficient and
linear regression.To illustrate the relationship between the two variables we could use a scatter plot with a regression line.The correlation 
coefficient value between Age and Experience is 0.99.A scatter plot was also showcased with Age as X axis and Experience as Y axis wherein 
the fit line: y = 24.990339 + 0.99x."""


```

```python
### 15. Do you see any outliers in the dataset? If yes, what plot you would think will be suitable to showcase to the stakeholders?

    
import pandas as pd


df = pd.read_csv('Bank Personal Loan Modelling (1).csv')


print(df.head())
print(df.info())


numeric_cols = ['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']


from scipy import stats


z_scores = stats.zscore(df[numeric_cols])
abs_z_scores = np.abs(z_scores)
outliers = (abs_z_scores > 3)  # Using a threshold of 3 for outliers
print(df[(outliers).any(axis=1)])  # Displaying the rows with outliers




Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1


outlier_mask = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))
outliers = df[outlier_mask.any(axis=1)]
print(outliers)  

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(14, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()












```

```python
### 17. Give the IQR of all the variables which are quantitative and continuous.


import pandas as pd


df = pd.read_csv('Bank Personal Loan Modelling (1).csv')


quant_cont_vars = ['Age', 'Experience', 'Income', 'CCAvg']


iqr_dict = {}
for var in quant_cont_vars:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    iqr_dict[var] = IQR


for var, iqr in iqr_dict.items():
    print(f"The IQR for {var} is {iqr:.2f}")






```

```python
### 18. Do the higher-income holders spend more on credit cards?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm


df = pd.read_csv('Bank Personal Loan Modelling (1).csv')


print(df[['Income', 'CCAvg']].isnull().sum())


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='CCAvg', data=df)
plt.title('Scatter Plot of Income vs Credit Card Spending')
plt.xlabel('Income')
plt.ylabel('Credit Card Spending (CCAvg)')
plt.show()


correlation = df['Income'].corr(df['CCAvg'])
print(f'Pearson correlation coefficient between Income and CCAvg: {correlation:.2f}')


X = df['Income']
y = df['CCAvg']
X = sm.add_constant(X)  # Add a constant term to the predictor


model = sm.OLS(y, X).fit()


print(model.summary())










```

```python
### 20. Using the z-score of the income variable, find out the number of observations outside the +-3.


import pandas as pd
import numpy as np

df = pd.read_csv('Bank Personal Loan Modelling (1).csv')


income_mean = df['Income'].mean()
income_std = df['Income'].std()

df['Income_z'] = (df['Income'] - income_mean) / income_std


outliers = df[(df['Income_z'] < -3) | (df['Income_z'] > 3)]
num_outliers = outliers.shape[0]

print(f"Number of observations outside the 3 range: {num_outliers}")



```
